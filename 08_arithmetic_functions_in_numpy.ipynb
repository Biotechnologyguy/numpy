```python
import numpy as np
```

## Arithmetic Functions

```python
# creating 1 dimensional arrays


num_arr = np.array([1,2,6,33,2,6,8,10])
```

### max()

```python
# Max

print(np.max(num_arr))
```

### argmax()

```python
# argmax - position of max

print(np.argmax(num_arr))
```

### min()

```python
# Min

print(np.min(num_arr))
```

### argmin()

```python
# argmin : position of min

print(np.argmin(num_arr))
```

### axis

```python
# axis 0 = row
# axis 1 = column
# by passing the axis, we can get mean for each axis 
print(np.min(num_arr , axis=0))

# TO GET THE Row of MIN
print(np.min([[1,3,99,5], [4,5,6,22]], axis=1))
print(np.min([[1,2,3,5], [4,2,2,2]], axis=0))
```

### mean()

```python
# Mean

print(np.mean(num_arr))
```

### sqrt()

```python
# Square root

print(np.sqrt(np.array([9, 25, 100, 64])))
```

### sin()

```python
print(np.sin(num_arr))
```

### cumsum()

```python
# we require it to find out median

print(np.cumsum(np.array([1,2,3,4])))
```
