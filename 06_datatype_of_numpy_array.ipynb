```python
my_arr = np.array([1,2,3,4])
print(f"Data type : {my_arr.dtype}")
```

```python
my_arr2 = np.array([1,2,3,4,5.0])
print(f"Data type : {my_arr2.dtype}")
```

```python
str_arr = np.array("Shubham")
print(f"Data type : {str_arr.dtype}")
```

```python
mix_arr = np.array(["Shubham", 1,2,3])
print(f"Data type : {mix_arr.dtype}")
```

```python
# How to change the data type 

my_arr = np.array([1,2,3,4])
print(f"Data type : {my_arr.dtype}")
my_arr = np.array([1,2,3,4], dtype=np.int8)
print(f"Data type after changing: {my_arr.dtype}")

```

```python
# Changing datatype to float

new_arr = np.float64(my_arr)
print(f"Data type new_arr: {new_arr.dtype}")
```
